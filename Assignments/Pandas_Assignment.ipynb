***PART 4: Students Performance***

***Tasks:***

1. Display average marks in math, reading, and writing.

```python
df3 = pd.read_csv('Datasets ML/StudentsPerformance.csv')
df3[['math score', 'reading score', 'writing score']].mean()
```

2. Who performed better in writing – males or females?

```python
df3.groupby('gender')['writing score'].mean()
```

3. Add a new column for total score and percentage.

```python
df3['total_score'] = df3['math score'] + df3['reading score'] + df3['writing score']
df3['percentage'] = (df3['total_score'] / 300) * 100
df3[['math score', 'reading score', 'writing score', 'total_score', 'percentage']].head()
```

4. Plot a bar chart showing average scores grouped by parental level of education.

```python
import matplotlib.pyplot as plt

df3['total_score'] = df3['math score'] + df3['reading score'] + df3['writing score']
avg_scores = df3.groupby('parental level of education')['total_score'].mean()
avg_scores.plot(kind='bar', color='lightblue')
plt.title('Average Total Score by Parental Education')
plt.ylabel('Average Total Score')
plt.xlabel('Parental Education')
plt.xticks(rotation=45)
plt.show()
```

5. Find students who scored above 90% in all subjects.

```python
df3[(df3['math score'] > 90) & 
         (df3['reading score'] > 90) & 
         (df3['writing score'] > 90)]
```
